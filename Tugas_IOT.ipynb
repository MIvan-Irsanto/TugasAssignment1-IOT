```python
import numpy as np
import pandas as pd
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```python
# Load the gyroscope data from a file or some other source
df = pd.read_csv('data_1600_accel.csv', sep=';', header=None)
df
```

```python
df.head()
```

```python
df.drop(0, axis='columns', inplace=True)
```

```python
df.head()
```

```python
df1 = df.drop([0])
```

```python
df1 = df1.apply(pd.to_numeric)
df1
```

```python
#df1 = df1.astype('int')
```

```python
# Set the sampling frequency based on the phone's gyroscope
fs = 100  # Hz

# Set the desired cutoff frequency and order for the filter
cutoff = 5  # Hz
order = 4    # filter order

# Generate the filter coefficients using the Butterworth filter
nyq = 0.5 * fs
normal_cutoff = cutoff / nyq
b, a = signal.butter(order, normal_cutoff, btype='lowpass', analog=False)

# Apply the filter to the gyroscope data
filtered_data = signal.filtfilt(b, a, df1, method='gust')

# Plot the original and filtered signals
t = np.arange(len(df1)) / fs
plt.plot(t, df1, 'b-', label='original')
plt.plot(t, filtered_data, 'g-', linewidth=2, label='filtered')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.title('Acceleration along X-axis')
plt.grid()
plt.legend()
plt.show()
```
